Initialisations

```python
import igraph

nodes = 97
tt = igraph.Graph()
tt.add_vertices(nodes)

length = tt.vs.indices
tt.add_vertices(3)
tt.add_edges([(0,1),(0,2),(0,3)])
#print(length)
used_nodes = [1,2,3]

length = tt.vs.indices
length.remove(0)
#print(length)
#type(length)
```

Telephone Tree Function

```python
def telephone_network(length):
    for vertice in length:
        tt.add_edges([(vertice, vertice + len(used_nodes))])
        used_nodes.append(vertice + len(used_nodes))
        tt.add_edges([(vertice, vertice + len(used_nodes))])
        used_nodes.append(vertice + len(used_nodes))
        used_nodes.remove(vertice)
        #print(used_nodes)
        if vertice == (nodes - 1) / 2:
            break
            
telephone_network(length)
```

Graph Layout

```python
tt.vs ["color"] = "white"
tt.es ["color"] = "red"
tt.vs ["label"] = [i for i in range (1, nodes + 4)]
#print(len(length))
igraph.plot(tt, layout = tt.layout("kk"))
```
